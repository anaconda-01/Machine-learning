```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Groceries_dataset.csv')
df.head()
```

```python
df2=pd.crosstab(df['Member_number'],df['itemDescription'])
df2
```

```python
df2[df2>0]=1
```

```python
df2
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
matrix=apriori(df2,min_support=0.02,use_colnames=True)

rules=association_rules(matrix,metric='lift',min_threshold=1)
```
