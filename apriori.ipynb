```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Groceries_dataset.csv')
df.head()
```

```python
df['ticket']=df['Member_number'].astype(str)
```

```python
df2=pd.crosstab(df['ticket'],df['itemDescription'])
df2[df2>0]=1
df2
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemset=apriori(df2,min_support=0.02,use_colnames=True)
rules=association_rules(frequent_itemset,metric='lift',min_threshold=1)
rules.head()
```

```python
rules.sort_values(['support','confidence','lift'],ascending=False,axis=0)
```
